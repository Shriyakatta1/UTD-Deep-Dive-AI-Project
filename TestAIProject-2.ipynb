Loading all the Temperature data from the last 5 years (only includes Texas Counties)

Creating the Temperature dataset


```python
Temperature25 = pd.read_csv("/content/avg_temps_2025.csv", header=2, skiprows=[2])
Temperature25.rename(columns={"Name": "County"}, inplace=True)
Temperature25['Year'] = 2025
print(Temperature25.shape)
#print(Temperature25.head())


Temperature24 = pd.read_csv("/content/avg_temps_2024.csv", header=2, skiprows=[2])
Temperature24.rename(columns={"Name": "County"}, inplace=True)
Temperature24['Year'] = 2024
print(Temperature24.shape)
#print(Temperature24.head())

Temperature23 = pd.read_csv("/content/avg_temps_2023.csv", header=2, skiprows=[2])
Temperature23.rename(columns={"Name": "County"}, inplace=True)
Temperature23['Year'] = 2023
print(Temperature23.shape)
#print(Temperature23.head())

Temperature22 = pd.read_csv("/content/avg_temps_2022.csv", header=2, skiprows=[2])
Temperature22.rename(columns={"Name": "County"}, inplace=True)
Temperature22['Year'] = 2022
print(Temperature22.shape)
#print(Temperature22.head())

Temperature21 = pd.read_csv("/content/avg_temps_2021.csv", header=2, skiprows=[2])
Temperature21.rename(columns={"Name": "County"}, inplace=True)
Temperature21['Year'] = 2021
print(Temperature21.shape)
#print(Temperature21.head())

Temperature = pd.concat([Temperature21, Temperature22, Temperature23, Temperature24, Temperature25])
Temperature['County'].value_counts()
#Temperature.columns
```

```python
AggTemp = Temperature.groupby('County')['Value'].agg(
    Mean='mean',
    Median='median',
    STD='std',
    Min='min',
    Max='max'
).reset_index()

AggTemp.rename(columns = {
    'Mean': 'Mean_Temp',
    'Median': 'Median_Temp',
    'STD': 'STD_Temp',
    'Min': 'Min_Temp',
    'Max': 'Max_Temp'
}, inplace = True)
AggTemp.rename(columns={'County': 'Name'}, inplace=True)
AggTemp['Mean_Temp'].round(2)
display(AggTemp)
```

```python
AggTemp.to_csv("Temperature_Aggregation.csv")
```
