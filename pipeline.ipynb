```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
import pandas as pd
from xgboost import XGBRegressor, XGBClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV
import t
```

```python
data = pd.read_csv('Combined_dataset_model.csv')
X = data.drop(['health_nor', 'temp_norm', 'biome', 'county'], axis = 1)
y1 = data['health_nor']
y2 = data['temp_norm']
```

```python
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size = 0.2, random_state = 42)
```

```python
param_grid = {
    'max_depth': [2, 3, 4, 5, 6, 7, 8, 9, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.5, 1]
}
grid_search = GridSearchCV(
    estimator = XGBRegressor(n_estimators = 100, enable_categorical = True, random_state=42),
    param_grid = param_grid,
    scoring='r2'
)
searched1 = grid_search.fit(X_train, y1_train)
print("y1 parameters:", searched1.best_params_)
print("y1 test accuracy:", round(searched1.best_estimator_.score(X_test, y1_test), 4))

searched2 = grid_search.fit(X_train, y2_train)
print("y2 parameters:", searched2.best_params_)
print("y2 test accuracy:", round(searched2.best_estimator_.score(X_test, y2_test), 4))
```

```python
xgb_model1 = XGBRegressor(n_estimators=100, max_depth=7, learning_rate=0.1, objective='reg:squarederror', enable_categorical = True)
xgb_model1.fit(X_train, y1_train)
xgb_pred1 = xgb_model1.predict(X_test)
print("R2 Score:", r2_score(y1_test, xgb_pred1))
print("MAE:", mean_absolute_error(y1_test, xgb_pred1))

xgb_model2 = XGBRegressor(n_estimators=100, max_depth=8, learning_rate=0.1, objective='reg:squarederror', enable_categorical = True)
xgb_model2.fit(X_train, y2_train)
xgb_pred2 = xgb_model2.predict(X_test)
print("R2 Score:", r2_score(y1_test, xgb_pred1))
print("MAE:", mean_absolute_error(y1_test, xgb_pred1))
```
