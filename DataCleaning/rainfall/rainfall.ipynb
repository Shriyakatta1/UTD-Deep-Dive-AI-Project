```python
import pandas as pd
```

```python
one = pd.read_csv('avg_rains_2021.csv', header = 2, skiprows = [2])
two = pd.read_csv('avg_rains_2022.csv', header = 2, skiprows = [2])
three = pd.read_csv('avg_rains_2023.csv', header = 2, skiprows = [2])
four = pd.read_csv('avg_rains_2024.csv', header = 2, skiprows = [2])
five = pd.read_csv('avg_rains_2025.csv', header = 2, skiprows = [2])
rains = pd.concat([one, two, three, four, five])
rains
```

```python
sorted_rain = rains.sort_values(by = ['Name'])
sorted_rain
```

```python
AggRain = pd.DataFrame()
AggRain = sorted_rain.groupby('Name')['Value'].mean().reset_index(name = 'Mean_Rain')
std = sorted_rain.groupby('Name')['Value'].std().reset_index(name = 'STD_Rain')
med = sorted_rain.groupby('Name')['Value'].median().reset_index(name = "Median_Rain")
AggRain = pd.concat([AggRain, med['Median_Rain']], axis = 1)
AggRain = pd.concat([AggRain, std['STD_Rain']], axis=1)
min = sorted_rain.groupby('Name')['Value'].min().reset_index(name = "Min_Rain")
AggRain = pd.concat([AggRain, min['Min_Rain']], axis = 1)
max = sorted_rain.groupby('Name')['Value'].max().reset_index(name = "Max_Rain")
AggRain = pd.concat([AggRain, max['Max_Rain']], axis = 1)

AggRain
```

```python
AggRain.to_csv('rainfall_cleaned.csv', index = False)
```
